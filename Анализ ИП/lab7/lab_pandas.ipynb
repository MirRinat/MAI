## 3 Постройте запрос: Владелец аэропорта, Аэропорт,Пассажиропоток суммарный за 2018 год, Грузопоток суммарный за 2018 год

```python
#Обработка df_airport
df_airport['Наименование аэропорта РФ']=df_airport['Наименование аэропорта РФ'].str.split().str.get(0)
# df_airport['Наименование аэропорта РФ']=df_airport['Наименование аэропорта РФ'].str.split('(').str.get(0)
```

```python
df_airport.head()
```

```python
#суммарный пассажиропоток
def replace_none(x):
    if type(x)==float:
        return x
    if x == '***':
        return None
    x = x.replace(' ','') 
    x = x.replace(',','.') 
    return float(x)

column_list_month = list(df_passenger)
del(column_list_month[0:2])
del(column_list_month[-1])
def preprocessing(df):

    
    df['Наименование аэропорта РФ']=df['Наименование аэропорта РФ'].str.split().str.get(0)
    
    for col in column_list_month:
        df[col]=df[col].apply(replace_none)
    df['Январь - Декабрь']=df[column_list_month].sum(axis=1)

preprocessing(df_passenger)
preprocessing(df_cargo)
df_passenger
```

```python
df_airport_owner = df_airport[["Наименование аэропорта РФ", "Владелец"]]
df_airport_owner 
```

```python
def select_df(df, text):
#     df['Год периода данных']=df['Год периода данных']
    df_sum = df[df['Год периода данных']==2018][['Наименование аэропорта РФ', 'Январь - Декабрь']]
    df_sum = df_sum[df_sum['Январь - Декабрь'] != 0.0]
    df_sum.rename({'Январь - Декабрь':text},axis=1,inplace=True)
    return(df_sum)

df_passenger_sum=select_df(df_passenger,"Суммарный пассажиропоток 2018")
df_passenger_sum
```

```python
df_cargo['Год периода данных']=df_cargo['Год периода данных'].apply(lambda x:x.replace(' ','')).astype(int)
df_cargo_sum = select_df(df_cargo,"Суммарный грузопоток 2018")


df_cargo_sum
```

```python
#merge
result = df_airport_owner.merge(df_passenger_sum,on="Наименование аэропорта РФ", how='outer')
result
```

```python
result = result.merge(df_cargo_sum,on="Наименование аэропорта РФ", how='outer')
result
```

## 4 Перечислите аэропорты, где пассажиропоток меньше медианы, а грузопоток больше медианы

```python
less_median_pass=df_passenger_sum[(df_passenger_sum['Суммарный пассажиропоток 2018']<df_passenger_sum['Суммарный пассажиропоток 2018'].median())]
less_median_pass
```

```python
more_median_cargo = df_cargo_sum[(df_cargo_sum['Суммарный грузопоток 2018']>df_cargo_sum['Суммарный грузопоток 2018'].median())]
more_median_cargo
```

```python
res = pd.merge(less_median_pass,more_median_cargo,on="Наименование аэропорта РФ",how="inner")
res
```

## 5 Перечислите авиакомпании у которых нет типов воздушных судов зарегистрированных в России 

```python
russian_aircraft=df_aircraft['Тип (наименование) воздушного судна'].unique()
russian_aircraft
```

```python
df_airlines
```

```python
def no_rus(all_craft):
    list_craft = all_craft.split(',')
    for x in list_craft:
        x = x.split('(')[0].rstrip()
        if x in russian_aircraft:
            return False
    return True

df_airlines[df_airlines['Гараж'].astype('str').apply(no_rus)]['Наименование']
```
